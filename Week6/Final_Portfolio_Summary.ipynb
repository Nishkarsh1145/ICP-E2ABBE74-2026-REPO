## Project 1 Summary: Binary Classification

## Project 2 Summary: Sentiment Analysis
- **Dataset:** IMDB Movie Reviews (50,000 reviews)
- **Best Model:** Bidirectional LSTM (TensorFlow/Keras)
- **Key Result:** ~90%+ accuracy on test set
- **Key Learning:** Deep learning captures sequential context in text better than bag-of-words

```python
import matplotlib.pyplot as plt
import numpy as np

# Final comparison chart across all models
models = [
    'LR\n(Binary)', 'DT\n(Binary)',
    'LR\n(Sentiment)', 'NB\n(Sentiment)', 'LSTM\n(Sentiment)'
]
accuracies = [0.97, 0.95, 0.89, 0.85, 0.91]
colors = ['#2196F3', '#2196F3', '#4CAF50', '#4CAF50', '#4CAF50']

plt.figure(figsize=(10, 5))
bars = plt.bar(models, accuracies, color=colors, width=0.5)
plt.ylim(0.7, 1.0)
plt.title('Final Model Accuracy Comparison — All Projects', fontsize=14)
plt.ylabel('Accuracy')
plt.xlabel('Model')

for bar, acc in zip(bars, accuracies):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.005,
             f'{acc:.0%}', ha='center', va='bottom', fontweight='bold')

# Legend
from matplotlib.patches import Patch
legend = [
    Patch(color='#2196F3', label='Project 1: Binary Classification'),
    Patch(color='#4CAF50', label='Project 2: Sentiment Analysis')
]
plt.legend(handles=legend)
plt.tight_layout()
plt.show()
```
